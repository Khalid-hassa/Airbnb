Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.pandas.set_option('display.max_columns',None)
```

```python
# importing dataset
data=pd.read_csv('datasets.csv',encoding_errors='ignore')
```

```python
# see the first 5 rows of data
data.head()
```

```python
# show the shape(number of col and rows) of the data
data.shape
```

```python
# show information about the data
data.info()
```

```python
# get basic statistics of numerical columns
data.describe()
```

```python
# check the null values in every column
data.isnull().sum()

# deletes the null value
data.dropna(inplace=True)

# after deleting the null values
print(data.isnull().sum())
```

```python
# again check the shape
data.shape
```

```python
# checking for duplicated values
data[data.duplicated()]

# drop the duplicate values
data.drop_duplicates(inplace=True)

data.shape
```

```python
# shows the data types of all the col
data.dtypes

# convert the [id and host_id] from int to object
data[['id','host_id']]=data[['id','host_id']].astype(object)

# convert the [last_review] col to date format
data['last_review']=pd.to_datetime(data['last_review'])

print(data.dtypes)
```

Checking Outliers for "price" column

```python
# find outlier using boxplot
sns.boxplot(data=data, x='price')
```

The boxplot shows that the dataset has outliers in price column. Most of the price lies below 10000.

```python
# only take the data where price < 1500
df=data[data['price'] < 1500]

# use boxplot to see price distribution
plt.figure(figsize=(8,6))
sns.boxplot(data=df, x='price')
```

The 2nd box plot shows the avg price lies between 100-200.

Price Distribution using Histogram

```python
plt.figure(figsize=(8,5))
sns.histplot(data=df, x='price',bins=50)
plt.ylabel('frequency')
plt.title('Distribution of prices')
plt.show()
```

From the histogram price distribution clearly reflects that the average price value has the max frequency coun

```python
# calculate the average_neighbourhood_price
average_neighbourhood_price=df.groupby('neighbourhood_group')['price'].mean()
print(average_neighbourhood_price)
```

```python
# calculate the average price for different room type
room_type_avg_price=df.groupby('room_type')['price'].mean().reset_index()

sns.barplot(data= room_type_avg_price, x='room_type',y='price',palette='viridis')
plt.title('Average Price by Room Type')
plt.xlabel('Room Type')
plt.ylabel('Average Price')
```

Hotel room, room type has the highest average price

```python
# check which feature impact price using heatmap correlation
num_cols=["price", "minimum_nights", "number_of_reviews", "availability_365"]
plt.figure(figsize=(8,5))
sns.heatmap(df[num_cols].corr(),annot=True)
plt.title("Feature Correlation with Price")
plt.show()
```

```python
# calculate avg_neighbourhood_price
avg_neighbourhood_price=df.groupby('neighbourhood')['price'].mean().sort_values(ascending=False).head(10).reset_index()

plt.figure(figsize=(14,6))
sns.barplot(data=avg_neighbourhood_price,x='neighbourhood',y='price',palette='coolwarm')
plt.title('Average Price by Neighbourhood')
plt.xlabel('Neighbourhood')
plt.ylabel('Average Price')
plt.show()
```

Top 10 Average price by neighbourhood

```python
# does number of reviews effect price
plt.figure(figsize=(10,6))
sns.scatterplot(x=df['price'],y=df['number_of_reviews'],hue=df['room_type'],alpha=0.6)
plt.title("Price vs. Review Score Across Neighborhoods")
plt.xlabel("Price")
plt.ylabel("Review Score")
plt.show()
```

Scatter plot reflects that the private_room, room_type has highest review. So host should focus on providing more features and services on private_room

```python
# create a new column month
df['month']=df['last_review'].dt.month
```

```python
# alculate the average price per month to see seasonal variations.
monthly_price=df.groupby('month')['price'].mean()

plt.figure(figsize=(10,6))
sns.lineplot(x=monthly_price.index,y=monthly_price.values,marker='o',color='r')
plt.title('Average Price by Month')
plt.xlabel('Month')
plt.ylabel('Average Price')
plt.grid()
plt.show()
```

There is a huge peak from month 1 to 2 and upto month 4 due to holidays

```python
monthly_price=df.groupby('month')['price'].mean()
monthly_demand=df.groupby('month')['number_of_reviews'].sum()

monthly_revenue=monthly_price*monthly_demand

# identify the best revenue month
plt.figure(figsize=(10,6))
sns.barplot(x=monthly_revenue.index,y=monthly_revenue.values,palette='coolwarm')
plt.title('Estimated Airbnb Revenue by Month')
plt.xlabel('Month')
plt.ylabel('Estimated Revenue')
plt.show()
```

After 7th month the revenue is continously increases and last month has the highest revenue. One of the main reasone can, people go out more in the month of december due to Christmas
